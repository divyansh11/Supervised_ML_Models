# Linear Regression

```python
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
```

```python
from sklearn.datasets import load_diabetes
```

```python
diab=load_diabetes()
```

```python
iris=load_iris()
```

```python
df_diab=pd.DataFrame(data=diab.data,columns=diab.feature_names)
```

```python
df_iris=pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
plt.scatter(df_iris['sepal length (cm)'],df_iris['sepal width (cm)'],c='red')
```

```python
df_diab.corr(method='pearson')
```

```python
x=df_iris.drop('petal width (cm)',axis=1)
y= df_iris['petal width (cm)']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
import seaborn as sns
corr=x_train.corr()  #pearson coffecient ranges form -1 to +1
```

```python
sns.heatmap(corr,annot=True,cmap=plt.cm.Accent)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
