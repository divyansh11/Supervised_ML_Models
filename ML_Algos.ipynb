# Linear Regression

```python
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
```

```python
from sklearn.datasets import load_diabetes
```

```python
diab=load_diabetes()
```

```python
iris=load_iris()
```

```python
df_diab=pd.DataFrame(data=diab.data,columns=diab.feature_names)
```

```python
df_iris=pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
plt.scatter(df_iris['sepal length (cm)'],df_iris['sepal width (cm)'],c='red')
```

```python
df_diab.corr(method='pearson')
```

```python
x=df_iris.drop('petal width (cm)',axis=1)
y= df_iris['petal width (cm)']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
import seaborn as sns
corr=x_train.corr()  #pearson coffecient ranges form -1 to +1
```

```python
sns.heatmap(corr,annot=True,cmap=plt.cm.Accent)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
from sklearn.tree import export_graphviz  
  
# export the decision tree to a tree.dot file 
# for visualizing the plot easily anywhere 
dot_data =export_graphviz(model, out_file ='tree.dot', 
               feature_names =['sepal length (cm)','sepal width (cm)','petal width (cm)'])  
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn import svm
model=svm.SVR(C=2,kernel='linear')
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
model=KNeighborsRegressor()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
```

```python
model_1=AdaBoostRegressor()
model_1.fit(x_train,y_train)
y_pred1=model_1.predict(x_test)
```

```python
model_1=GradientBoostingRegressor()
model_1.fit(x_train,y_train)
y_pred2=model_1.predict(x_test)
```

```python
model_1=XGBRegressor()
model_1.fit(x_train,y_train)
y_pred3=model_1.predict(x_test)
```

```python
model_1=LGBMRegressor()
model_1.fit(x_train,y_train)
y_pred4=model_1.predict(x_test)
```

```python
model_1=CatBoostRegressor()
model_1.fit(x_train,y_train)
y_pred5=model_1.predict(x_test)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(figsize=(6, 4))
 
ax = sns.lineplot(x=y_test, y=y_pred1,
                  label='GradientBoosting')
ax1 = sns.lineplot(x=y_test, y=y_pred2,
                   label='XGBoost')
ax2 = sns.lineplot(x=y_test, y=y_pred3,
                   label='AdaBoost')
ax3 = sns.lineplot(x=y_test, y=y_pred4,
                   label='CatBoost')
ax4 = sns.lineplot(x=y_test, y=y_pred5,
                   label='LightGBM')
 
ax.set_xlabel('y_test', color='g')
ax.set_ylabel('y_pred', color='g')
```

# Classification Problem

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('train-news.csv')
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df=df.fillna("null")
```

```python
df.head()
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df['content']=df['title']+df['author']
```

```python
x=df['content'].values
y=df['label'].values
```

```python
x
```

```python
vectorizer=TfidfVectorizer()
vectorizer.fit(x)
```

```python
x=vectorizer.transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)

```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(x_test)
```
