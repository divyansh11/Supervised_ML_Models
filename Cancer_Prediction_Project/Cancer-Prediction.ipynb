```python
def correlation(dataset,threshold):
    s=set()
    for i in range(len(dataset.columns)):
        for j in range(i):
            if abs(dataset.iloc[i, j]) > threshold:
                colname_i = dataset.columns[i]
                colname_j = dataset.columns[j]
                if i > j:
                    s.add(colname_i)
                else:
                    s.add(colname_j)
                    

    return s
    
    
```

```python
#Assigning correleation thereshold, suuch that if the vaue is greater than a thereshold, then drop those columns
correlation(x_corr,0.9)
```
